## Prac w9 - ensemble methods

Fit a single decision tree classifier to the training data with bagging and boosting ensemble methods for classification

```python
import pandas as pd 
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.inspection import DecisionBoundaryDisplay
from matplotlib.colors import ListedColormap
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
w3_classif = pd.read_csv('w3classif.csv', header=None)
w3_classif = w3_classif.rename(columns={0: "x1", 1: "x2", 2:'y'})

print(w3_classif.head())

#testing and training split
from sklearn.model_selection import train_test_split

X = w3_classif[['x1', 'x2']]
y = w3_classif['y']

X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.30, random_state=42)

cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])
cmap_bold_list = ['#FF0000', '#0000FF']
cmap_light_list = ['#FFAAAA', '#AAFFAA', '#AAAAFF']

clf = DecisionTreeClassifier().fit(X_train, y_train)

# Plot the decision boundary
DecisionBoundaryDisplay.from_estimator(
    clf,
    X,
    cmap=cmap_light,
    response_method="predict",
    xlabel='x',
    ylabel='y',
)

# Plot the training points
sns.scatterplot(
    x=X_train['x1'],
    y=X_train['x2'],
    hue=y,
    palette=cmap_bold_list,
    alpha=1.0,
    edgecolor="black",
)

# Plot the training points
sns.scatterplot(
    x=X_test['x1'],
    y=X_test['x2'],
    hue=y,
    palette=cmap_light,
    alpha=1.0,
    edgecolor="black",
)

plt.title(
    'Decision tree classifier'
    )

plt.show()
```

## Bagging ensemble method

```python
from sklearn.ensemble import RandomForestClassifier

clf_rand_forest = RandomForestClassifier().fit(X_train, y_train)

DecisionBoundaryDisplay.from_estimator(
    clf_rand_forest,
    X,
    cmap=cmap_light,
    response_method="predict",
    xlabel='x',
    ylabel='y',
)

# Plot both the training and the testing points
sns.scatterplot(
    x=X['x1'],
    y=X['x2'],
    hue=y,
    palette=cmap_bold_list,
    alpha=1.0,
    edgecolor="black",
)

plt.title(
    'Decision tree classifier'
    )

plt.show()
```
